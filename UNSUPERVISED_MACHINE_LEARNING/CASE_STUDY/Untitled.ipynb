```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
pd.options.display.max_rows= None
pd.options.display.max_columns= None
```

```python
df=pd.read_csv('data.csv')
df.head()
```

```python
df=df.drop(['id','diagnosis'],axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
q1=df.quantile(0.25)
q3=df.quantile(0.75)
iqr=q3-q1
ll=q1-1.5*iqr
ul=q3+1.5*iqr
df_trim=df[~((df<ll)|(df>ul)).any(axis=1)]
```

```python
df_trim.describe()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df_trim_sc=sc.fit_transform(df_trim)
df_trim_sc=pd.DataFrame(df_trim_sc,columns=df_trim.columns)
df_trim_sc.head()
```

```python
from sklearn.cluster import KMeans
kmns=KMeans(n_clusters=2,random_state=10)
# fit the model
kmns.fit(df_trim_sc)
```

```python
cluster_range=range(1,15)
wcss=[]
cluster_sil_score=[]
for i in range(1,21):
    kmeans=KMeans(n_clusters=i,random_state=10)
    kmeans.fit(df_trim_sc)     # cluster or centroid form
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,21),wcss,marker='o')
plt.axvline(x=2,color='red')
plt.title('Elbow Plot', fontsize=15)
plt.xlabel('No. of Cluster (k)', fontsize=10)
plt.ylabel('WCSS',fontsize=10)
```

```python
from sklearn.metrics import silhouette_score
n_clusters= [2,3,4,5,6,7,8,9,10]
for k in n_clusters:
    cluster=KMeans(n_clusters=k,random_state=10)
    predict=cluster.fit_predict(df_trim_sc)
    score=silhouette_score(df_trim_sc,predict,random_state=10)
    print('for {} cluster the silhouette score is {})' .format(k,score))
```

```python
from yellowbrick.cluster import SilhouetteVisualizer
```

```python
for i in range(2,10):
    kms=KMeans(n_clusters=i)
    kms.fit(df_trim_sc)
    model= KMeans(n_clusters=i,random_state=42)
    visualizer= SilhouetteVisualizer(model)
    visualizer.fit(df_trim_sc)
    visualizer.show()
```
