# IRIS DATA SET 

```python
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
df1=datasets.load_iris()
```

```python
df1
```

```python
x=df1.data
```

```python
df_x= pd.DataFrame(x,columns=['sapal-length','sepal-width','petal-length','petal-width'])
```

```python
df_x.head()
```

```python
df_x.shape
```

```python
sns.pairplot(df_x,diag_kind='kde')
```

```python
sns.heatmap(df_x.corr(),annot=True)
```

```python
sc=StandardScaler()     # clustering,PCA(principle component analysis) required scaling 
```

```python
x_sc=sc.fit_transform(x)
```

```python
cov_matrix= np.cov(x_sc.T)
cov_matrix
# Covariance will be 4*4 Matrix as there are 4 features
```

```python
eigval, eigvec =np.linalg.eig(cov_matrix)
```

```python
print('Eigen Vectors \n',eigvec)
print('\nEigen Values \n',eigval)
```

```python
print('The first eigen vector is:')
counter = 0
for i in range(0,len(eigvec[0])):
    counter = counter+1
    if(counter != (len(eigvec[0]))):
        print(eigvec[i,0],'*',df_x.columns[i],"+")
    else:
        print(eigvec[i,0],'*',df_x.columns[i])
        
```

```python
tot=sum(eigval)
tot
```

```python
var_exp = [(i/tot)*100  for i in sorted(eigval,reverse=True)]
var_exp
```

```python
cum_var_exp =np.cumsum(var_exp)
cum_var_exp
```

```python
plt.plot(eigval,'bp')            

# plot a line plot
plt.plot(eigval) 

plt.title('Scree Plot', fontsize = 15)
plt.xlabel('Principal Components', fontsize = 15)    
plt.ylabel('Explained Variance', fontsize = 15)  
plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
pca_iris=PCA().fit(x_sc)
```

```python
pca_iris
```

```python
pca_iris.explained_variance_      # eigen value
```

```python
pca_iris.components_      # eigen vector
```

```python
eigvector=pca_iris.components_
eigvector[0,0]
```

```python
print('The first eigen vector is:')
counter = 0
for i in range(0,len(eigvector[0])):
    counter = counter+1
    if(counter != (len(eigvector[0]))):
        print(eigvec[i,0],'*',df_x.columns[i],"+")
    else:
        print(eigvec[i,0],'*',df_x.columns[i])
        
```

```python
pca_iris.n_components_
```

```python
pca_iris.explained_variance_ratio_
# For example, let's say we have performed PCA on a dataset and obtained three principal components, 
# and the explained variance ratios of these components are [0.5, 0.3, 0.2]. This means that the first principal component 
# explains 50% of the variance in the original data, the second principal component explains 30% of the variance, and the 
# third principal component explains 20% of the variance.
```

```python
var=np.cumsum(np.round(pca_iris.explained_variance_ratio_,decimals=4)*100)
var
```

```python
pca_iris_new = PCA(n_components=2)
data_reduced = pca_iris_new.fit_transform(x_sc)
data_reduced.transpose()
```

```python
pca_df= pd.DataFrame(data_reduced)
pca_df.head()
```

```python
pca_df.columns=['PC0','PC1',]
pca_df.head()
```

```python
corr = pca_df.corr().round(decimals=4)
corr
```

```python
sns.heatmap(corr,annot=True, fmt='.2f',cmap='CMRmap',vmin=1,vmax=1)
plt.show()
```

# LINEAR DISCRIMINANT ANALYSIS(LDA)

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
y=df1.target
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(x_sc,y,test_size=0.2,random_state=10,stratify=y)
```

```python
iris_lda=LDA()
iris_lda_df=iris_lda.fit(X_train,y_train)
```

```python
iris_lda_df.explained_variance_ratio_
```

```python
iris_lda_df=iris_lda.fit_transform(X_train,y_train)
iris_lda_df
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
iris_lda.transform(X_test)
```

```python
y_pred_lda=iris_lda.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred_lda))
print(accuracy_score(y_test,y_pred_lda))
```
