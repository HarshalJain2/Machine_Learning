```python
# h0: there is no significant difference in age between stroke. 
# ha: there is significant difference in age between stroke..

stroke_df = df[df["stroke"] == 1]
no_stroke_df = df[df["stroke"] == 0]

ttest_result = stats.ttest_ind(stroke_df["age"], no_stroke_df["age"])

print("T-statistic: ", ttest_result.statistic)
print("P-value: ", ttest_result.pvalue)
```

```python
# The results of the t-test you performed on the age variable indicate a very low p-value of 7.03e-71, 
# which is less than any reasonable significance level such as 0.05. This suggests that there is a significant difference in 
# age between stroke and non-stroke patients, with the stroke group being generally older than the non-stroke group. 
# The t-statistic value of 18.08 further supports this conclusion, indicating that the difference in age between 
# the two groups is very large relative to the variability within each group.
```
